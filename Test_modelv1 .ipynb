```python
import tensorflow as tf
```

```python
model = tf.keras.models.load_model('Final models/model_v1.keras')

model.summary()
```

```python
from PIL import Image
import numpy as np

def reshape_image(image_path, target_size=(96, 96)):
    image = Image.open(image_path)

    # Ensure the image is in RGB mode
    if image.mode != 'RGB':
        image = image.convert('RGB')

    resized_image = image.resize(target_size)
    
    image_array = np.array(resized_image)

    return image_array

def prepare_image_for_prediction(image_array):

    expanded_image_array = np.expand_dims(image_array, axis=0)
    
    normalized_image_array = expanded_image_array / 255.0
    
    return normalized_image_array
```

```python
image_path = 'Testing images/angry nek.jpg'
reshaped_image = reshape_image(image_path)

prepared_image = prepare_image_for_prediction(reshaped_image)

prediction = model.predict(prepared_image)


predicted_class = np.argmax(prediction)
predicted_class_map = {0: 'angry',
                       1: 'disgust',
                       2: 'fear',
                       3: 'happy',
                       4: 'sad',
                       5: 'Surprise',
                       6: 'Neutral'}

print(f'The predicted class is: {predicted_class_map[predicted_class]}')
```
